## Generate Pandas Scatter Plots

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Median vs Sample Size')
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Unemployment Rate vs Sample Size')
```

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='Unemployment Rate vs Share Women')
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Median vs Men')
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Median vs Women')
```

```python
recent_grads.plot(x='ShareWomen', y='Median', kind='scatter', title='Median vs Share Women')
```

### Findings
#### â€¢Students in more popular majors do NOT make more money.
#### â€¢From the plot Median vs Share Women students who majored in subjects that were majority female did NOT make more money. The plot Unemployment Rate vs Share Women also coroborates this showing that unemployment rates are higher in fields where there are more women.
#### â€¢There is a positive correlation between the number of full-time employees and median salary. Fields with full-time employees have higher median salaries.
